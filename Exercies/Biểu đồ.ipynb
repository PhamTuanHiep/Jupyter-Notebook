# 1. Đồ thị
### Vẽ đồ thị hàm số :  $f(x)=x^2+x+1$

```python
def f(x):
    return x**2 + x + 1
# Make data
x_1 = np.linspace(start=-3,stop=3,num=500)
#plot
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1))
plt.text(-0.5,1, 'cực tiểu')
plt.show()
```

# 2.Biểu đồ tròn ( Pie Chart)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Apple','Banana','Chery','Lemon', 'Guava'])
y = np.array([100,200,500,300,600])
myexplode = [0,0,0,0.2,0]
mycolor = ['Pink','green','red','blue','yellow']

plt.pie(y, labels=x, explode = myexplode, colors = mycolor, shadow = True, autopct = '% 2.3f  %%')
plt.legend()
plt.show()
```

# 3.Biểu đồ đường ( Line Chart)

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1,3,8,9])
y_points = np.array([10,24,15,50])
plt.plot(x_points, y_points, lw='10', linestyle='dashed',color ='r')
plt.show()
```

# 4, Biểu đồ cột ( barchar)

```python
food = ['eggs','meat','fish','vegetable']
food_consume=[30,2.5,3,8]
plt.bar(food, food_consume, width=0.5 , color= 'magenta')
plt.ylabel('consume')
plt.title('consume/week')
plt.xlabel('food')
plt.text(1,5,'hihi')
# plt.legend()
plt.show()
```

```python
import numpy as np

men_means, men_std = (20, 35, 30, 35, 27), (2, 3, 4, 1, 2)
women_means, women_std = (25, 32, 34, 20, 25), (3, 5, 2, 3, 3)

ind = np.arange(len(men_means))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, men_means, width, yerr=men_std,
                label='Men')
rects2 = ax.bar(ind + width/2, women_means, width, yerr=women_std,
                label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()

plt.show()
```

```python
# Import Library

import numpy as np
import matplotlib.pyplot as plt

# Create figure and subplots

fig, ax = plt.subplots()
   
# Define Data Coordinates

x = np.linspace(0, 20 , 100)
y = np.sin(x)

# Plot

plt.plot(x, y)

# Set ticks

ax.set_xticks(np.arange(0, len(x)+1, 5))

# Add Title
 
fig.suptitle('set_xticks Example', fontweight ="bold")
               
# Display
               
plt.show()
```
